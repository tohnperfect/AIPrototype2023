```python
path = "/content/drive/MyDrive/wolf_AI/full data/TechConBiz Data/Techconsbiz"
```

```python
all_folders = os.listdir(path)
```

```python
save_path = "/content/drive/MyDrive/wolf_AI/tables/"
```

```python
all_main = []

for e_doc in all_folders:
    all_main.append(pd.read_csv(os.path.join(save_path,f'main_{e_doc}.csv')))

main_df = pd.concat(all_main)
```

```python
all_hist = []

for e_doc in all_folders:
    all_hist.append(pd.read_csv(os.path.join(save_path,f'hist_{e_doc}.csv')))

hist_df = pd.concat(all_hist)
```
